Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Converting the Dataset from text format to csv format and saving the files

```python
rows = []
file_name = ['train_data', 'test_data_solution']
with open(file_name[1]+'.txt', 'r') as train_dataset:
  for line in train_dataset:
    line.strip()
    id, title, genre, description = line.split(':::')
    rows.append({'Id': id, 'Title': title, 'Genre': genre, 'Description': description})

df = pd.DataFrame(rows)
df.to_csv(file_name[1]+'.csv', index=False)
```

```python
rows = []

with open('test_data.txt', 'r') as test_dataset:
  for line in test_dataset:
    line.strip()
    id, title, description = line.split(':::')
    rows.append({'Id': id, 'Title': title, 'Description': description})

df = pd.DataFrame(rows)
df.to_csv('test_data.csv', index=False)
```

```python
# Checking if dataset has missing values
print(df.isnull().sum())
```

```python
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```
